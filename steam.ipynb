**Steam** is a <ins>video game digital distribution</ins> service and storefront from <ins>Valve</ins>. Steam offers various features, like <ins>digital rights management</ins> (DRM), <ins>game server matchmaking</ins> with <ins>Valve Anti-Cheat</ins> measures, <ins>social networking</ins>, and <ins>game streaming</ins> services.
